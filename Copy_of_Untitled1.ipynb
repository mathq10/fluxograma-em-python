```python
import random
import matplotlib.pyplot as plt

# Simulação de tráfego
def simular_trafego(tempo_total=10):
    tempo = []
    fluxo_normal = []
    fluxo_otimizado = []

    for t in range(1, tempo_total + 1):
        tempo.append(t * 20)
        fluxo_normal.append(t * random.randint(4, 6))
        fluxo_otimizado.append(t * random.randint(5, 8))

    return tempo, fluxo_normal, fluxo_otimizado

# Simulação de consumo de energia
def simular_consumo_energia(horas=24):
    residencial = [random.randint(50, 80) if 6 <= h <= 22 else random.randint(20, 40) for h in range(horas)]
    comercial = [random.randint(80, 120) if 8 <= h <= 18 else random.randint(10, 30) for h in range(horas)]
    return residencial, comercial

# Plotagem
def plotar_resultados():
    tempo, normal, otimizado = simular_trafego()
    residencial, comercial = simular_consumo_energia()

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(tempo, normal, label='Fluxo Normal', color='magenta')
    plt.plot(tempo, otimizado, label='Fluxo Otimizado', color='orange')
    plt.title('Simulação de Tráfego')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Número de Veículos')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(residencial, label='Residencial', color='blue')
    plt.plot(comercial, label='Comercial', color='green')
    plt.title('Consumo de Energia (24h)')
    plt.xlabel('Hora')
    plt.ylabel('Consumo (kWh)')
    plt.legend()

    plt.tight_layout()
    plt.show()

plotar_resultados()
```
